```python
#Random under sampling
count_class_0,count_class_1 = df.Class.value_counts()
df_class_0 = df[df['Class']==0]
df_class_1 = df[df['Class']==1]
df_class_0_under = df_class_0.sample(count_class_1)
df_under = pd.concat([df_class_0_under,df_class_1],axis=0)
df_under.Class.value_counts().plot(kind='bar', title='Count under sampling')
```

```python
#Random over sampling
df_class_1_over = df_class_1.sample(count_class_0,replace=True)
df_over = pd.concat([df_class_0,df_class_1_over],axis=0)
df_over.Class.value_counts().plot(kind='bar', title='Count over sampling')
```

```python
#Split data into train and test sets

X_under = df_under.drop('Class', axis=1)
y_under = df_under['Class']
X_under_train, X_under_test, y_under_train, y_under_test = train_test_split(X_under,y_under, test_size=0.25,random_state = 2)
print('Number of training examples in under-sampled dataset: ', len(X_under_train))
print('Number of test examples in under-sampled dataset: ', len(X_under_test))

X_over = df_over.drop('Class', axis=1)
y_over = df_over['Class']
X_over_train, X_over_test, y_over_train, y_over_test = train_test_split(X_over,y_over, test_size=0.25,random_state = 2)
print('Number of training examples in over-sampled dataset: ', len(X_over_train))
print('Number of test examples in over-sampled dataset: ', len(X_over_test))
```
